```python
import pandas as pd
```

```python
for i in range(len(data_df)):
    if data_df.at[i, 'rating'] > 3:
        data_df.at[i, 'rating'] = 1
    else:
        data_df.at[i, 'rating'] = 0
    if i % 100000 == 0:
        print i
```

```python
data_df['userId'] = data_df['userId'] + 1
data_df['itemId'] = data_df['itemId'] + 1
```

```python
df = data_df.copy()
```

```python
df.drop(df.index[df['rating'] == 0], axis=0, inplace=True)
df
```

```python
df.drop('user_freq', axis=1, inplace=True)
df.drop('item_freq', axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
import numpy as np
user_table = np.zeros(np.max(df['userId']) + 1)
movie_table = np.zeros(np.max(df['itemId']) + 1)
user_set = set(df['userId'].tolist())
movie_set = set(df['itemId'].tolist())
print('start')
u = 1
for i in range(1, np.max(df['userId']) + 1):
    if i in user_set:
        user_table[i] = u
        u += 1
    if i % 20000 == 0:
        print('u' + str(i))
m = 1
for i in range(1, np.max(df['itemId']) + 1):
    if i in movie_set:
        movie_table[i] = m
        m += 1
    if i % 10000 == 0:
        print('m' + str(i))
```

```python
print u
print m
```

```python
tmp = df.values
for i in range(len(df)):
    tmp[i, 0] = user_table[int(tmp[i, 0])]
    tmp[i, 1] = movie_table[int(tmp[i, 1])]
    if i % 100000 == 0:
        print i
df = pd.DataFrame(tmp, columns=['userId', 'itemId'])
df
```

```python
num_user = u - 1
num_movie = m - 1
print('number of users = ' + str(num_user))
print('number of items = ' + str(num_movie))
sparsity = len(df) * 1.0 / (num_user * num_movie)
print('sparsity = ' + str(sparsity))
```

```python
df.to_csv('./data.csv', index=False)
```

```python
train_df = df.copy()
test_df = df.copy()
```

```python
train_ratio = 0.8
test_ratio = 1 - train_ratio
num_users = np.max(df['userId'])
num_items = np.max(df['itemId'])
test_idx = []
for u in range(1, num_users+1):
    u_idx = train_df.index[train_df['userId'] == u]
    idx_len = len(u_idx)
    test_len = int(idx_len * test_ratio)
    if test_len == 0:
        test_len = 1
    tmp = np.random.choice(u_idx, size=test_len, replace=False)
    test_idx += tmp.tolist()
    if u % 500 == 0:
        print u
```

```python
test_set = set(test_idx)
train_set = set(range(len(df)))
train_set -= test_set
train_idx = list(train_set)
train_df.drop(test_idx, axis=0, inplace=True)
test_df.drop(train_idx, axis=0, inplace=True)
```
